````python
# Desenvolva seu código aqui
def verificaPalindromo(palavra):
    palavra_reversed = palavra[::-1]
    if palavra_reversed == palavra:
      print(f"A palavra {palavra} eh um palindromo")
    else:
      print(f"A palavra {palavra} Nao eh palindromo")


verificaPalindromo("arara")
````
