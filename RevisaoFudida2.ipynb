## Questão 1 (Dificuldade 1)

**Enunciado**: Escreva uma função que recebe uma string e retorna a quantidade de vogais na string.

Exemplo de Entrada: `'ciencia de dados'`

Saída Esperada: `'A string "ciencia de dados" tem 7 vogais.'`

**Dica**: Lembre de contar ignorando se ela está em maiúscula ou minúscula para este exercício. Exemplo: `a` e `A` para este exercício devem ser contadas como iguais. Uma opção é transformar todos os caracteres para minúsculo usando o método `lower()` na string.

````python
# Desenvolva seu código aqui
def contaVogal(frase):
  #uso a variavel contador como um placeholder (google explica melhor a parada do placeholder)
  contador = 0

  #aqui eu itero sobre a string pegando todas as letras (usando o metodo .lower() para transformar tudo em minusculo)
  for i in frase.lower():
    #uma condicional para comparar as letras iteradas com a sequencia aeiou que representa as vogais
    if i in "aeiou":
      #contador para contar o numero de vogais que ocorrem na string, cada vez q tiver vogal ele vai adicionar 1 unidade
      contador +=1
  print(f"O numero de vogais da string '{frase}' sao {contador}")

contaVogal("Alberani nao tem cabelo")
````

**Enunciado**: Implemente uma função que recebe uma lista de números e retorna uma nova lista contendo apenas os números pares.

Exemplo de Entrada: `[1, 2, 3, 4, 5, 6]`

Saída Esperada: `'Números pares: [2, 4, 6]'`


````python
# Desenvolva seu código aqui

def contaPares(lista):
  #crio uma lista vazia que vai receber os valores pares do lista digitada pelo usuario
  lista_pares = []

  #itero sobre a lista digitada pelo usuario
  for i in lista:

    #uso a condicional para achar apenas os numeros pares (se i % 2 e o resto for 0 entao ele eh par)
    if i % 2 ==0:

      #e adiciono a lista_pares
      lista_pares.append(i)
  print(f"A lista com os pares ficou: {lista_pares}")

contaPares([1,4,3,7,512,13])






````

## Questão 3 (Dificuldade 3)

**Enunciado**: Crie uma função que verifica se uma palavra é um palíndromo (palavras que lêem o mesmo de trás pra frente).

Exemplo de Entrada: `'arara'`

Saída Esperada: `'A palavra "arara" é um palíndromo.'`

**Dica**: use a função `range(tamanho_lista // 2)` para iterar com índice comparando a primeira metade com a segunda metade

````python
# Desenvolva seu código aqui
def verificaPalindromo(palavra):
  #crio uma string com a palavra inversa
    palavra_reversed = palavra[::-1]

    #nesse if eu apenas comparo a palavra inversa com a palavra digitada pelo usuario
    if palavra_reversed == palavra:
      #se ela for igual eh um palindromo
      print(f"A palavra {palavra} eh um palindromo")
    else:
      print(f"A palavra {palavra} Nao eh palindromo")


verificaPalindromo("matheus")
````

Saída Esperada: `'Alunos aprovados: {'Ana': 8, 'Carla': 9}'`

**Dica**: para iterar (fazer um loop) em um dict, use seu método `items()`.
Este método retorna duas variáveis, a chave e o valor.

Exemplo para o dict chamado `notas_alunos`:

```python

  ...código do loop...
```

````python
# Desenvolva seu código aqui
def aprova_aluno(dicionario):
  #aqui eu crio um dicionario vazio que no final vai ser ele que vai ser retornado
  dicio_aprovados =  {

  }
  #aqui eu itero (novamente iterar) sobre as keys e values do dicionario (chaves e valores).
  for key,value in dicionario.items():

    #aqui eu verifico qual dos valores eh maior ou igual a 7
    if value>=7:

      #e assim eu termino adicionando as chaves ao dicionario vazio de acordo com o dicionario passado, e botando apenas os alunos acima de media 7
      dicio_aprovados[key] = value

  print(f"Os Alunos aprovados sao {dicio_aprovados}")
dicionario_exemplo = {
    'matheus':6,
    'calvo':9,
    'dodoi':8
}

aprova_aluno(dicionario_exemplo)
````

**Enunciado**: Implemente uma função que recebe um número inteiro e retorna a soma dos dígitos desse número.

A função deve verificar se o número é positivo e inteiro. Se o número fornecido não for positivo ou não for inteiro, a função deve lançar uma exceção `ValueError`. Após a execução da função, caso a soma dos dígitos seja bem-sucedida, exiba uma mensagem amigável com o resultado. Se uma exceção ValueError for capturada, exiba uma mensagem amigável informando o usuário sobre o erro.

Exemplo de Entrada 1: `123`

Saída Esperada 1: `'A soma dos dígitos é 6.'`

Exemplo de Entrada 2: `-1`

Saída Esperada 2: `'Erro: O valor deve ser um número inteiro positivo.'`

````python
# Desenvolva seu código aqui
def soma_digitos(numero):
  try:
    #inicialmente eu transformo o numero (que voce digitou quando chama a funcao) em uma string e tu ja vai entender o porque
    numero = str(numero)

    #crio a variavel soma como um placeholder (qualquer coisa pesquisa no google q ele consegue explicar melhor)
    soma = 0

    #esse looping vai iterar (tbm eh bom tu pesquisar pra ficar familiarizado) sobre cada digito do numero que vc usuario colocou
    for digito in numero:
      #aqui eu estou transformando cada digito do numero em inteiro pq se nao nao da pra realizar a operacao
      digito = int(digito)

      #lembra da soma? entao, a cada digito que ele passar no numero que tu digitou, ele vai atribuir a variavel soma
      soma +=digito

    print(f"A soma dos digitos '{numero}' = {soma}")

  #essa parte eh pra basicamente avisar o bizonho que colocou o numero errado
  except ValueError:
    print("Erro, o valor deve ser um numero inteiro positivo")

soma_digitos(123)

````
